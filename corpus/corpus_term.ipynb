```python
w_uniq = list(set(words))
w_uniq.sort()
w_uniq.index('when') #37
w_uniq.index('à¦à¦²à¦¬à§') #54

w_uniq = w_uniq[:38] + w_uniq[54:]
```

```python
def wCounter(data, data_u):
    cnt = []
    for w in data_u:
        c = data.count(w)
        cnt.append(c)
    return cnt
```

```python
num_s = []
num_s = wCounter(words, w_uniq)
```

```python
from more_itertools import sort_together
def wSort(data_n, data_u):
    num = []
    words = []
    num = list(sort_together([data_n, data_u])[0])
    words = list(sort_together([data_n, data_u])[1])

    order = []
    order.append(num[::-1])
    order.append(words[::-1])
    o = np.array(order).T
    return o
```

```python
timingOrder = wSort(num_s, w_uniq)
```
