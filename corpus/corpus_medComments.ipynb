```python
d = df['d']
d = d.values.tolist()
d
```

```python
words = []
s = []
l = []

for i in range(len(d)):
    if not isinstance(d[i], float):
        s = str(d[i])
        l = s.replace(",", " ").replace("(", " ").replace(")", " ").replace("#", " ").replace("*", " ").replace(".", " ").replace("&", " ").replace(">", " ").split()
        words.extend(l)
```

```python
cnt = 1
while cnt != 0:
    cnt = 0
    for x in range(len(words)):
        if len(words[x]) != 0:
            if words[x][0] == '+' or words[x][0] == '-' or words[x][0] == '/':
                cnt = 1
                words[x] = words[x][1:]
```

```python
w_uniq = list(set(words))
w_uniq.sort()
```

```python
w_uniq
```

```python
def wCounter(data, data_u):
    cnt = []
    for w in data_u:
        c = data.count(w)
        cnt.append(c)
    return cnt
```

```python
num_s = []
num_s = wCounter(words, w_uniq)
```

```python
num_s
```

```python
from more_itertools import sort_together
def wSort(data_n, data_u):
    num = []
    words = []
    num = list(sort_together([data_n, data_u])[0])
    words = list(sort_together([data_n, data_u])[1])

    order = []
    order.append(num[::-1])
    order.append(words[::-1])
    o = np.array(order).T
    return o
```

```python
comOrder = wSort(num_s, w_uniq)
```

```python
comOrder
```

```python
comOrderList = []
comOrderList = comOrder.tolist()
comOrderList
```

```python
from langdetect import detect

bn = []
en = []
ot = []
for x in range(len(comOrderList)):
    try:
        language = detect(comOrderList[x][1])
        if language == 'bn':
            bn.append(comOrderList[x])
        elif language == 'en':
            en.append(comOrderList[x])
    except:
        language = "error"
        ot.append(comOrderList[x])
```
