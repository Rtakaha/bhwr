```python
b = df['b']
b = b.values.tolist()
b
```

```python
kind = []
name = []
dose = []
meal = []
s = []

for x in range(len(b)):
    if not isinstance(b[x], float):
        if len(b[x]) > 10:
            s = [w.strip() for w in b[x].split("-")]
            kind.append(s[0])
            name.append(s[1])
            dose.append(s[2])
            meal.append(s[3])
```

```python
kind_u = list(set(kind))
name_u = list(set(name))
dose_u = list(set(dose))
meal_u = list(set(meal))

kind_u.sort()
name_u.sort()
dose_u.sort()
meal_u.sort()
```

```python
len(name_u)
```

```python
def wCounter(data, data_u):
    cnt = []
    for w in data_u:
        c = data.count(w)
        cnt.append(c)
    return cnt
```

```python
num_k = []
num_n = []
num_d = []
num_m = []

num_k = wCounter(kind, kind_u)
num_n = wCounter(name, name_u)
num_d = wCounter(dose, dose_u)
num_m = wCounter(meal, meal_u)
```

```python
from more_itertools import sort_together
def wSort(data_n, data_u):
    num = []
    words = []
    num = list(sort_together([data_n, data_u])[0])
    words = list(sort_together([data_n, data_u])[1])

    order = []
    order.append(num[::-1])
    order.append(words[::-1])
    o = np.array(order).T
    return o
```

```python
kOrder = wSort(num_k, kind_u)
nOrder = wSort(num_n, name_u)
dOrder = wSort(num_d, dose_u)
mOrder = wSort(num_m, meal_u)
```

```python
import pickle
with open("csv/separateCorpus/kOrder.txt", "wb") as fp:   #Pickling
    pickle.dump(kOrder, fp)
with open("csv/separateCorpus/nOrder.txt", "wb") as fp:   #Pickling
    pickle.dump(nOrder, fp)
with open("csv/separateCorpus/dOrder.txt", "wb") as fp:   #Pickling
    pickle.dump(dOrder, fp)
with open("csv/separateCorpus/mOrder.txt", "wb") as fp:   #Pickling
    pickle.dump(mOrder, fp)
```
