```python
g = df['g']
g = g.values.tolist()
g
```

```python
words = []
s = []
l = []

for i in range(len(g)):
    if not isinstance(g[i], float):
        s = str(g[i])
        l = s.replace(",", " ").replace("(", " ").replace(")", " ").replace("#", " ").replace("*", " ").replace(".", " ").replace("&", " ").replace(">", " ").split()
        words.extend(l)
```

```python
w_uniq = list(set(words))
w_uniq.sort()
w_uniq
```

```python
def wCounter(data, data_u):
    cnt = []
    for w in data_u:
        c = data.count(w)
        cnt.append(c)
    return cnt
```

```python
num_s = []
num_s = wCounter(words, w_uniq)
```

```python
from more_itertools import sort_together
def wSort(data_n, data_u):
    num = []
    words = []
    num = list(sort_together([data_n, data_u])[0])
    words = list(sort_together([data_n, data_u])[1])

    order = []
    order.append(num[::-1])
    order.append(words[::-1])
    o = np.array(order).T
    return o
```

```python
Order = wSort(num_s, w_uniq)
```
