```python
a = []
b = []
c = []
d = []
e = []
f = []
g = []
test = []

# column num "a" ~ "g" まで
a = df[df['a'].notnull()]
b = df[df['b'].notnull()]
c = df[df['c'].notnull()]
d = df[df['d'].notnull()]
e = df[df['e'].notnull()]
f = df[df['f'].notnull()]
g = df[df['g'].notnull()]

test = a.values.tolist()
test.extend(b.values.tolist())
test.extend(c.values.tolist())
test.extend(d.values.tolist())
test.extend(e.values.tolist())
test.extend(f.values.tolist())
test.extend(g.values.tolist())
```

```python
test # セルごとのセンテンス
```

```python
len(test[0])
```

```python
words = []
s = []
l = []

for i in range(len(test)):
    for j in range(len(test[i])):
        s = str(test[i][j])
        l = s.replace(",", " ").replace("(", " ").replace(")", " ").replace("#", " ").replace("*", " ").replace(".", " ").replace("&", " ").replace(">", " ").split()
        words.extend(l)
```

```python
cnt = 1
while cnt != 0:
    cnt = 0
    for x in range(len(words)):
        if len(words[x]) != 0:
            if words[x][0] == '+' or words[x][0] == '-' or words[x][0] == '/':
                cnt = 1
                words[x] = words[x][1:]
```
